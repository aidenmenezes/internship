The objective of this assignment was to design a conceptual Spam Detection System. The task involved identifying possible features used for spam detection, understanding what data is required, and analyzing the potential mistakes such systems might make.

Spam classifiers are commonly used in email services to automatically filter unwanted messages.


Possible Features for Spam Detection

Presence of suspicious words (e.g., Free, Win, Offer, Urgent)

Number of links in the email

Sender email reputation

Message length

Use of capital letters or special characters

Data Required

To build a spam classifier, a dataset of emails labeled as:

Spam (1) – Unwanted promotional or malicious emails

Not Spam (0) – Legitimate emails

The dataset should include the email text, sender information, and labels indicating whether the message is spam or not.

Possible Mistakes (Model Errors)

False Positive: A normal email is incorrectly marked as spam.

False Negative: A spam email is incorrectly classified as normal.

Misclassification due to new spam words not present in training data.

Emails with mixed content may confuse the model.
