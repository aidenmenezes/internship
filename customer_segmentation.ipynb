```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = {
    "Age": [19,21,20,23,31,22,35,23,64,30,67,35,58,24,37],
    "AnnualIncome": [15,15,16,16,17,17,18,18,19,19,20,20,21,21,22],
    "SpendingScore": [39,81,6,77,40,76,6,94,3,72,14,99,15,77,13]
}

df = pd.DataFrame(data)

kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(df[["AnnualIncome","SpendingScore"]])

plt.figure()
plt.scatter(df["AnnualIncome"], df["SpendingScore"], c=df["Cluster"])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()

print("Cluster Centers:")
print(kmeans.cluster_centers_)
```
