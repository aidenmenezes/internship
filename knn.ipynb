Netflix-like Recommendation Using KNN

Suppose we have ratings for movies given by users.

User	Action	Comedy	Drama
User1	5	1	2
User2	4	2	1
User3	1	5	4
User4	2	4	5

If a new user likes Action movies highly, the system finds users with similar preferences using distance calculations. Movies liked by those similar users are then recommended.

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors
data = np.array([
    [5,1,2],  
    [4,2,1],  
    [1,5,4],  
    [2,4,5]   
])

knn = NearestNeighbors(n_neighbors=2, metric='euclidean')
knn.fit(data)
new_user = np.array([[5,1,1]])

distances, indices = knn.kneighbors(new_user)

print("Most similar users:", indices)
print("Distances:", distances)
```
