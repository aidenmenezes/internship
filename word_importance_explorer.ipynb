```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [
    "Machine learning is very powerful",
    "AI and machine learning are related",
    "Deep learning is a subset of machine learning",
    "AI is transforming the world",
    "Learning algorithms improve performance"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

features = vectorizer.get_feature_names_out()

for i, doc in enumerate(X):
    scores = zip(features, doc.toarray()[0])
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)
    print(f"\nTop words in Document {i+1}:")
    for word, score in sorted_scores[:3]:
        print(word, ":", round(score, 2))
```
