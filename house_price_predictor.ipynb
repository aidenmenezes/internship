```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
data = {
    "size_sqft": [500, 700, 900, 1100, 1300, 1500, 1700, 2000],
    "price": [100000, 140000, 180000, 220000, 260000, 300000, 340000, 400000]
}

df = pd.DataFrame(data)

X = df[["size_sqft"]]
y = df["price"]


model = LinearRegression()
model.fit(X, y)

new_house = pd.DataFrame({"size_sqft": [1200]})
predicted_price = model.predict(new_house)

print("Predicted Price for 1200 sqft house:", predicted_price[0])

plt.scatter(df["size_sqft"], df["price"])
plt.plot(df["size_sqft"], model.predict(X))
plt.xlabel("House Size (sqft)")
plt.ylabel("House Price")
plt.title("House Price Prediction")
plt.show()
```
